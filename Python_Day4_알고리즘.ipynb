# 선형 탐색 알고리즘 Linear Search

탐색 알고리즘이란, 원하는 데이터를 찾아내는 알고리즘이다.

대부분의 검색 엔진 naver, daum도 탐색 알고리즘을 사용한다.

- **제일 앞부터 순서대로 조사**하여 찾는 알고리즘

- 단순하여 이해가 쉬우나,

- 탐색 효율 자체는 그리 좋지 못하다.

```python
i = 0
arr = [4,2,3,5,1]

while i < 5:
  if arr[i] == 5:
    print(f"{i}번째에 있습니다.")
    break
  else:
    if i == 4:
      print("찾는 값이 없습니다.")
      break
    else:
      i = i + 1
```

# 단순 선택 정렬 알고리즘 Straight Selection Sort

가장 작은 데이터를 선택하여 맨 앞부터 순서대로 정렬한다.

image.png

image.png

```python
def selection_sort(arr):
    for i in range(len(arr) - 1):  # 리스트의 마지막 수는 어차피 자기 자신과 비교되므로 -1 해준다.
        min_idx = i  #가장 작은 수가 들어갈 인덱스 위치
        for j in range(i + 1, len(arr)): # 최소값이 있는 위치(arr[i])를 제외한 나머지 구간에서
            if arr[j] < arr[min_idx]:  # arr[j] < arr[i]와 하나씩 비교하며 arr[i]보다 작은 수를 찾는다.
                min_idx = j  # 작은 수로 최소 인덱스가 바뀌고
                t = arr[i] #j와 i 위치 바꿔주기
                arr[i] = arr[min_idx]
                arr[min_idx] = t
                #arr[i], arr[min_idx] = arr[min_idx], arr[i]

arr = [12,13,11,14,10]
selection_sort(arr)
print(arr)

# 그림을 다시 그려보자.
```
