```python
# 1. 벅스 크롤링 https://music.bugs.co.kr/chart

import requests   # 파이썬으로 인터넷 사이트 연결
from bs4 import BeautifulSoup as bs  # 인터넷 데이터 분석시 필수. 인터넷으로 긁어온 데이터 정제
```

# requests
파이썬에서 http를 사용하기 위해 쓰여지는 라이브러리

# BeautifulSoup
html 정보로부터 원하는 데이터를 가져오기 쉽게, 비슷한 분류의 데이터별로 나누어주는 파이썬 라이브러리.

```python
url = "https://music.bugs.co.kr/chart"
html = requests.get(url)

print(html)

# <Response [200]> : 읽어오기 성공 코드
```

```python
print(html.text)   # 읽어온 페이지 코드 보기

# 아래의 읽어온 데이터는 코드가 아닌 그냥 '글자'로 인식된다. 분석할 수 없다.
# 따라서 이것을 코드로 바꿔주기 위해 BeautifulSoup를 사용해 정제한다.
```

```python
soup = bs(html.text)

print(soup)

# 정제 완료.
# 위와 차이는 없어보이지만, 내부적으로 다르다.
```

```python
# 100등까지 뽑아내기

song_data = []
rank = 1   # 굳이 순위를 뽑을 필요 없이 여기서 설정해주자.

url = "https://music.bugs.co.kr/chart"   # 사이트 연결
html = requests.get(url)                 # 페이지 퍼오기
soup = bs(html.text)                     # 페이지 가공 by beautifulSoup
songs = soup.select("table.byChart > tbody > tr")   # 100곡 추출

for i in range(100):
    title = songs[i].select('p.title > a')[0].text
    singer = songs[i].select('p.artist > a')[0].text
    song_data.append(["Busgs", rank, title, singer])
    rank += 1

print(song_data)
```

```python
# 2

song_data = []
rank = 1   

url = "https://music.bugs.co.kr/chart"   
html = requests.get(url)                
soup = bs(html.text)                     
songs = soup.select("table.byChart > tbody > tr")  

for song in songs:
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    song_data.append(["Busgs", rank, title, singer])
    rank += 1



# 2리스트로 만든 100곡을 데이터프레임으로 전환하기
import pandas as pd
df = pd.DataFrame(song_data, columns = ['서비스','순위','노래제목','가수']) # 맨 위 인덱스
print(df)


# 데이터 프레임을 엑셀 파이롤 저장하기
df.to_excel('Bugs.xlsx', index=False)
```

위에서 살펴본 벅스 사이트는 보안이 없어서 쉽게 requests로 소스 코드를 가져올 수 있었지만,

멜론, 지니는 사이트를 못 퍼가게 보안 사항이 적용되어 막아뒀다.

따라서 대상 사이트에서 직접 코드를 읽어오는 방식이 아니라

우리에게 쏴주는 '웹 브라우저'의 코드를 읽어오는 우회 방법을 사용한다.


**크롬 드라이브 설치**

        ... -> 설정 -> 정보 -> 브라우저 버전 확인 -> 해당 버전을 설치 

        -> 압축 파일 풀어서 exe 파일을 작업하는 파일과 같은 위치로 이동

반드시 브라우저와 같은 버전의 드라이브를 설치해야 한다!!!



```python
from selenium import webdriver   # 브라우저 연결 / requests 대신
from bs4 import BeautifulSoup as bs
import pandas as pd

song_data = []
rank = 1  

driver = webdriver.Chrome()   # 크롬 드라이버 위치, 슬래시 방향 바꿔주기, 마지막에 크롬 드라이브 이름 추가
# driver 경로를 파일경로와 같은 곳에 둘 경우 이렇게 작성해도 된다.
# 다를 경우: 괄호 안에 드라이브의 경로를 넣어줘야한다. driver = webdriver.Chrome('C:/Users/EZEN/Documents/Python class/chromedriver.exe')


url = "https://www.melon.com/chart/index.htm"
driver.get(url)   # 드라이버로 가서 위 주소를 읽어온다

html = driver.page_source   # api를 이용해 현재 랜더링 된 페이지의 elements를 모두 가져온다.
# 브라우저에 보이는 그대로의 HTML, 크롬 개발자 도구의 Element 탭 내용과 동일.
# requests 를 통해 가져온 req.text 는 http 요청 결과로 받아온 html

soup = bs(html)
                          
songs = soup.select("tbody > tr")  

for song in songs:
    title = song.select('div.rank01 > span > a')[0].text    #클래스 명에 띄어쓰기가 있을 땐 뒤에 것만 써준다.
    singer = song.select('div.rank02 > span > a')[0].text
    song_data.append(["Melon", rank, title, singer])
    rank += 1

df = pd.DataFrame(song_data, columns = ['서비스','순위','노래제목','가수'])
print(df)
```
